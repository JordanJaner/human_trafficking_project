```python
import pandas as pd
import numpy as np
import os
```

```python
fips_file = "Resources/fips_for_ori_processing.csv"
ori_file = "Resources/ht_for_ori_processing.csv"
```

```python
ori_df = pd.read_csv(ori_file)
ori_df.head()
```

```python
ori_df['COUNTY']= ori_df['COUNTY_NAME'] + " " + ori_df['STATE_ABBR']
ori_df
```

```python
fips_df = pd.read_csv(fips_file)
fips_df.head()
```

```python
fips_df['Name']=fips_df['Name'].str.upper()
```

```python
fips_df
```

```python
fips_df['COUNTY']= fips_df['Name'] + " " + fips_df['State']
```

```python
fips_df
```

```python
fips_df['State'].unique()
```

```python
fips_df['State'].nunique()
```

```python
# fips_df['State'] = fips_df['State'][fips_df['State'].line_race == 'AS']
# fips_df.drop(fips_df['State'].index[fips_df['State']['line_race'] == 'AS'], inplace = True)
fips_df = fips_df[fips_df['State'] != 'AS']
fips_df = fips_df[fips_df['State'] != 'MP']
fips_df = fips_df[fips_df['State'] != 'GU']
fips_df = fips_df[fips_df['State'] != 'PR']
fips_df = fips_df[fips_df['State'] != 'VI']
```

```python
ori_short_df = ori_df[["DATA_YEAR",
                      "ORI",
                      "PUB_AGENCY_NAME",
                      "POPULATION_GROUP_CODE",
                      "POPULATION_GROUP_DESC",
                      "OFFENSE_SUBCAT_ID",
                      "OFFENSE_SUBCAT_NAME",
                      "ACTUAL_COUNT",
                      "COUNTY"]]
ori_short_df
```

```python
merge_df = pd.merge(ori_short_df, fips_df, on="COUNTY")
merge_df
```

```python
merge_short_df = merge_df[['DATA_YEAR', 'OFFENSE_SUBCAT_ID', 'OFFENSE_SUBCAT_NAME', 'ACTUAL_COUNT', 'COUNTY', 'FIPS', 'Name', 'State']]
merge_short_df
```
